```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img21=cv.imread('b.png',-1)
```

```python
fimg21=np.fft.fftshift(np.fft.fftn(img21))
```

```python
fimg212=np.log(np.abs(fimg21)**2)
```

```python
plt.imshow(np.log(np.abs(fimg21)**2))
```

```python
cordx=int(((np.shape(img21)[0]/2)+13))
```

```python
cordy=int(((np.shape(img21)[1]/2)+15))
```

```python
mascara=np.ones(np.shape(img21))
cv.circle(mascara,(cordx, cordy),8,0,-1)
plt.imshow(mascara)
plt.show
```

```python
cordy2=int(((np.shape(img21)[1]/2)-16))
cordx2=int(((np.shape(img21)[0]/2)-13))
mascara2=np.ones(np.shape(img21))
cv.circle(mascara2,(cordx2, cordy2),8,0,-1)
plt.imshow(mascara2)
plt.show
```

```python
mascaraf=mascara*mascara2
plt.imshow(mascaraf)
plt.show
```

```python
fourierplane=fimg212*mascaraf
plt.imshow(fourierplane)
plt.show
```
