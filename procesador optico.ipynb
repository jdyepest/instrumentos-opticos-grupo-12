```python
def procesador_optico_4f(entryimage,mask):
    fourierimage1=np.fft.fftshift(np.fft.fftn(entryimage)) #propaga la trasmitancia de entrada atravez de la primera lente hasta el plano de fourier
    plt.imshow(np.log(np.abs(fourierimage1)))#imagen con visualizacion
    visfourier=(np.log(np.abs(fourierimage1)))
    plt.show()
    fourieriamge2=fourierimage1*mask  #multiplica el campo propagado con la funcion mascara en el plano de fourier
    plt.imshow(np.abs(visfourier*mask))
    plt.show()
    finalimage=np.abs(np.fft.ifftn(fourieriamge2)) #repropaga la imagen atravez de la segunda lente y gira el plano para que la imagen quede de nuevo en la orientacion
    #adecuanda
    plt.imshow(finalimage)

    cv.imwrite('imagen result.png', finalimage)# conserva la imagen su analogo puede ser una pelicula fotografica
```

```python
procesador_optico_4f(imgmuestra,mask3)
```
